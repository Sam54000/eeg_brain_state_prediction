```python
import glob
import pandas as pd
import os
from pathlib import Path
root = '/data2/Projects/eeg_fmri_natview/derivatives/'
path = Path(root)
a = path.rglob()
file_system = {
    "root": [],
    "subject": [],
    "session": [],
    "datatype": [],
    "task": [],
    "run": [],
    "description": [],
    "suffix": [],
    "extension": [],
}
for file in a:
    file_system['root'].append(root)
    path_parts = file.relative_to(path).parts
    file_system['subject'].append(path_parts[0].split('-')[1])
    file_system['session'].append(path_parts[1].split('-')[1])
    file_system['datatype'].append(path_parts[2])
    desired_keys = ['task','run','desc','acq']
    for file_part in file.stem.split('_'):
        if len(file_part.split('-')) > 1:
            key, value = file_part.split('-') 
            if key in desired_keys:
                if key == 'desc':
                    key = 'description'
                elif key == 'acq':
                    key = 'acquisition'
                file_system[key].append(value)

    file_system['suffix'].append(file_part)
            
    file_system['extension'].append(file.suffix)

df = pd.DataFrame(file_system)

```
