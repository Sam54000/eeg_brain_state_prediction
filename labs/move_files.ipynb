```python
import bids_explorer.architecture.architecture as arch
import shutil
import os
from bids_explorer.utils.parsing import parse_bids_filename
import re
from pathlib import Path
source_folder = Path("/projects/EEG_FMRI/bids_eeg/BIDS/NEW/PREP_BVA_GR_CB_BK_NOV2024")
destination_folder = Path("/data2/Projects/eeg_fmri_natview/raw")
for file in source_folder.iterdir():
    entities = parse_bids_filename(file)
    if entities["description"] == "GdCbBk":
        destination_filename = file.name.replace("_desc-GdCbBk", "")
        destination_path = Path(destination_folder,
                                f"sub-{entities['subject']}",
                                f"ses-{entities['session']}",
                                "eeg",
                                destination_filename)
        if not destination_path.exists():
            destination_path.parent.mkdir(parents=True, exist_ok=True)
        
        print(f"Copying {file} to {destination_path}")
        shutil.copy(file, destination_path)
        
```
